```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
colors = colors
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(30,30))
for ax, col in zip(axes.flat, df.T.columns):
    ax.pie(df.T[col], labels=df.T.index, autopct='%.2f', colors=colors)
    ax.set(ylabel='', title=col, aspect='equal')

axes[0, 0].legend(bbox_to_anchor=(0, 0.5))
```
